## Data Visualization

### Import needed libraries

````r
library(repr)
````

### Import Data into ```data.frame```

#### Background data

````r
data.bkg = read.csv("../data/background.csv")
````

````r
dim(data.bkg)
````

#### Signal data

````r
data.sig = read.csv("../data/MC_signal.csv")
````

````r
dim(data.sig)
````

#### LHCb data

````r
data.lhcb = read.csv("../data/data_lhcb.csv")
````

````r
dim(data.lhcb)
````

### Check which features are present

````r
data.frame(names(data.bkg))
````

### Plot some features

#### $\Lambda_{b}^0$ mass

````r
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.bkg$Lambda_b0_MM_F, breaks=100, col="firebrick3", xlab="Lb0 Mass", main="Background")
hist(data.sig$Lambda_b0_MM_F, breaks=100, col="firebrick3", xlab="Lb0 Mass", main="Signal")
````

````r
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.lhcb$Lambda_b0_MM_F, breaks=100, col="firebrick3", xlab="Lb0 Mass", main="LHCb data")
````

#### $\Lambda_{c}^{+}$ mass

````r
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.bkg$Lambda_c_MM_F, breaks=100, col="firebrick3", xlab="Lc+ Mass", main="Background")
hist(data.sig$Lambda_c_MM_F, breaks=100, col="firebrick3", xlab="Lc+ Mass", main="Signal")
````

````r
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.lhcb$Lambda_c_MM_F, breaks=100, col="firebrick3", xlab="Lc+ Mass", main="LHCb data")
````

#### $\Delta M$

````r
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.bkg$DeltaM_F, breaks=100, col="firebrick3", xlab="Delta Mass", main="Background")
hist(data.sig$DeltaM_F, breaks=100, col="firebrick3", xlab="Delta Mass", main="Signal")
````

````r
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.lhcb$DeltaM_F, breaks=100, col="firebrick3", xlab="Delta Mass", main="LHCb data")
````

#### $\Lambda_{b}^0$ Decay Time

````r
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.bkg$lbDecTime_F[data.bkg$lbDecTime_F<10], breaks=100, col="firebrick3", xlab="Lb0 Decay Time", main="Background")
hist(data.sig$lbDecTime_F[data.sig$lbDecTime_F<10], breaks=100, col="firebrick3", xlab="Lb0 Decay Time", main="Signal")
````

````r
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.lhcb$lbDecTime_F[data.lhcb$lbDecTime_F<10], breaks=100, col="firebrick3", xlab="Lb0 Decay Time", main="LHCb data")
````

#### $\Lambda_{c}^{+}$ Decay Time

````r
data.bkg$lcDecTime_F
````

````r
par(mfrow=c(1,2))
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.bkg$lcDecTime_F[data.bkg$lcDecTime_F<1 & data.bkg$lcDecTime_F>-1], breaks=100, col="firebrick3", xlab="Lc+ Decay Time", main="Background")
hist(data.sig$lcDecTime_F[data.sig$lcDecTime_F<1 & data.sig$lcDecTime_F>-1], breaks=100, col="firebrick3", xlab="Lc+ Decay Time", main="Signal")
````

````r
options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
hist(data.lhcb$lcDecTime_F[data.lhcb$lcDecTime_F<1 & data.lhcb$lcDecTime_F>-1], breaks=100, col="firebrick3", xlab="Lc+ Decay Time", main="LHCb data")
````

### Plot all

````r
for (i in 1:dim(data.lhcb)[2]) {
    options(repr.plot.width=16, repr.plot.height=8, repr.plot.res=150)
    hist(data.lhcb[[i]], breaks=100, xlab=names(data.lhcb)[[i]], col="firebrick3", main="LHCb data")
}
````
